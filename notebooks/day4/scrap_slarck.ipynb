```
import time
from datetime import datetime
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
```

```
def parser(link):
  response = get(link)
  html_soup = BeautifulSoup(response.text, 'html.parser')
  return html_soup

def extract_info(html_soup):
  final_data = []
  temp_data = []
  temp_link = ""
  for key, data in enumerate(html_soup.findAll(True, {'class':['chat-msg', 'chat-msg-chrome']})):
    if data['class'][0] == "chat-msg-chrome":
      if len(temp_data) >= 3:
        temp_data.append(temp_link)
        final_data.append(temp_data)
        temp_link = ""
      member = data.find_all('span', {'class' : 'chat-msg-auth'})[0].text
      date_ = data.find_all('span', {'class' : 'chat-msg-ts'})[0]
      temp_data = [date_['title'],member]

    elif data['class'][0] == "chat-msg":
      text = data.find_all("div")
      link = data.find_all('a', href=True)
      if link:
        temp_link = temp_link + " " +link[0].text
        temp_link = temp_link.strip()
        if text:   
          text = data.find_all("div")[0].text.split(" ")
          if len(temp_data) >= 3:
            output_text = temp_data[-1] + " " + " ".join(text)
            temp_data[-1] = output_text
          else:
            temp_data.append(" ".join(text))
        else:
          temp_data.append("no data")
         
  return final_data

def create_dataframe(data):
  return pd.DataFrame(data, columns = ['date','username' , 'text', 'Link'])

                      
may_df = create_dataframe(extract_info(parser("https://slarck.com/4gXceH/showcase/2019/05/?k=oDnZML9eWIa72UizNVZZ7DFsIpzgBlQ3Tih0Wy5P")))
june_df = create_dataframe(extract_info(parser("https://slarck.com/4gXceH/showcase/2019/06/?k=oDnZML9eWIa72UizNVZZ7DFsIpzgBlQ3Tih0Wy5P")))
july_df = create_dataframe(extract_info(parser("https://slarck.com/4gXceH/showcase/2019/07/?k=oDnZML9eWIa72UizNVZZ7DFsIpzgBlQ3Tih0Wy5P")))
```

```
df = pd.concat([may_df,june_df, july_df], ignore_index=True)
```
