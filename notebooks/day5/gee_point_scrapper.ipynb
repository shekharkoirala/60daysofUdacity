```
import pandas as pd
import argparse

cols ={
        'coords_x' : 'coord_obs_x',
        'coords_y' :'coord_obs_y',
        'segment_id' :'segment_id',
        'district': 'district',
        'lc_group1' : 'lc_group1',
        'lc_code1' : 'lc_code1',
        'segment_id' : 'segment_id',
        'lc_pct1 ': 'lc_pct1',
        'irrig': 'irrig',
		'su_date':'su_date'
      }
color = ['ff0000','00ff00','0000ff','ffff00']

def main():
#     parser = argparse.ArgumentParser()
#     parser.add_argument("--path", help="Where is your CSV file?", dest='path') 
#     parser.set_defaults(path='./point_survey_v1.csv')
#     args = parser.parse_args()

#     path = args.path
    df = pd.read_csv('./point_survey_v1.csv')
    while True:
        query = input('What is the x coordinates? Include the "-" in your response.\n>')
        query ='.'.join(query.split('-'))
        found = df.loc[df[cols['coords_x']].astype(str) == query]
        segment = found.segment_id.values[0]
        found_segments =  df.loc[df[cols['segment_id']] == segment]
        results ={k: found_segments[cols[k]].values for k in cols.keys()}

        output = 'You may paste this in the code editor. \n\n'
        for n,(x,y) in enumerate(zip(results['coords_x'],results['coords_y'])):
            output += f'var point{n+1} = ee.Geometry.Point([{x}, {y}]); \n'
            output += f'Map.addLayer(point{n+1},{{"color": "{color[n]}"}}); \n'
        output += 'Map.centerObject(point1, 17);'

        for n in range(len(found_segments)):
            output += f'\n\nPoint {n+1} \n============================'
            for k in list(results.keys())[2:]:
                output += f'\n{k} : {results[k][n]}'
        print(output)
        print('\nLet\'s do another?\n')
```

```
main()
```
